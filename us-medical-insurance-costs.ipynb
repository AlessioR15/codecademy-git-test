# U.S. Medical Insurance Costs

```python
import csv

# Let's create empty lists with all the categories:

age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

# Let's write on this lists every data from insurance.csv

with open('insurance.csv', newline='') as database_ex:
    database_reader = csv.DictReader(database_ex)
    for info in database_reader:
        age.append(info['age'])
        sex.append(info['sex'])
        bmi.append(info['bmi'])
        children.append(info['children'])
        smoker.append(info['smoker'])
        region.append(info['region'])
        charges.append(info['charges'])
        
# Let's define a function that take all the lists in one dicitonary
        
def dictionary_data(age, sex, bmi, children, smoker, region, charges):
    database = dict()
    len_data = len(age)
    for x in range(len_data):
        database.update({x+1: {'age': age[x], 'sex': sex[x], 'bmi': bmi[x], 'children': children[x], 'smoker': smoker[x], 'region': region[x], 'charges': charges[x]}})
    return database
      
database = dictionary_data(age, sex, bmi, children, smoker, region, charges)

# Let's the challange begin...
#Â The first difference we want to discover is the one between southwest and northeast. 
# Are there some difference about the average cost in medical insurance costs?


def data_for_region(database):
    data_south_west = []
    data_south_east = []
    data_north_west = []
    data_north_east = []
    data_region = dict()
    for x,data in database.items():
        if data['region'] == "southwest":
            data_south_west.append(data['charges'])
        elif data['region'] == "southeast":
            data_south_east.append(data['charges'])
        elif data['region'] == "northeast":
            data_north_east.append(data['charges'])
        elif data['region'] == "northwest":
            data_north_west.append(data['charges'])
    data_region.update({'southwest' : data_south_west, 'southeast' : data_south_east, 'northeast' : data_north_east, 'northwest' : data_north_west})    
    return data_region

data_region = data_for_region(database)

def fix_number(number):
    numbers = 0
    for i in number:
        numbers = numbers + float(i)
    numbers = numbers / len(number)
    numbers = int(numbers)
    return numbers

data_r_s_w = data_region['southwest']
data_r_s_w = fix_number(data_r_s_w)

data_r_s_e = data_region['southeast']
data_r_s_e = fix_number(data_r_s_e)

data_r_n_w = data_region['northwest']
data_r_n_w = fix_number(data_r_n_w)

data_r_n_e = data_region['northeast']
data_r_n_e = fix_number(data_r_n_e)


total_by_region = {data_r_s_w: 'Southwest', data_r_s_e : 'Southeast', data_r_n_w: 'Northwest', data_r_n_e: 'Notheast'}
total_by_region = sorted(total_by_region.items())


print("The average insurance cost sorted by region: \n" + str(total_by_region[0]) + str(total_by_region[1]) + str(total_by_region[2]) + str(total_by_region[3]))


# Now let's calculate the difference between No Smoker and Smoker:

def smoker_difference(database):
    smoker = []
    no_smoker = []
    smokers = dict()
    for y,data in database.items():
        if data['smoker'] == "yes":
            smoker.append(data['charges'])
        elif data['smoker'] == "no":
            no_smoker.append(data['charges'])
    smokers.update({'smoker' : smoker, 'no smoker' : no_smoker})    
    return smokers

smokers = smoker_difference(database)

smoker_total = smokers['smoker']
smoker_total = fix_number(smoker_total)
no_smoker_total = smokers['no smoker']
no_smoker_total = fix_number(no_smoker_total)

print("\nIn terms of insurance costs, a smoker spends, on average: $" + str(smoker_total - no_smoker_total) + " more than a non-smoker.")

#Now let's calculate the average sort by sex


def sex_difference(database):
    male = []
    female = []
    sex = dict()
    for z,data in database.items():
        if data['sex'] == "male":
            male.append(data['charges'])
        elif data['sex'] == "female":
            female.append(data['charges'])
    sex.update({'male' : male, 'female' : female})    
    return sex

sex = sex_difference(database)
male = sex['male']
male = fix_number(male)

female = sex['female']
female = fix_number(female)


print("\nIn terms of insurance costs, a man spends, on average: $" + str(male - female) + " more than a woman.")

```
